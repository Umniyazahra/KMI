Pie Chart

```python

font2 = {'family': 'Arial',
        'color':  'magenta',
        'weight': 'bold',
        'size': 40,
        }

labels = df_can['major_category'].unique()
sizes = df_can.groupby(['major_category']).size()
colors = ['lightcoral', 'crimson', 'orchid', 'fuchsia', 'darkviolet', 'violet', 'hotpink', 'pink', 'lavender']

fig, ax = plt.subplots(figsize=(15, 10))
ax.set_facecolor("darkorchid")

plt.title('Percentage of Crimes in Major Category\n', fontdict=font2)
ab = plt.pie(sizes, labels=labels, colors=colors, startangle=90, autopct='%1.1f%%', textprops={'fontsize': 14, 'color': 'blueviolet'})

plt.legend(labels, loc="best", prop={'size': 10}, bbox_to_anchor=(1, 0.5))

plt.suptitle('Crime Categories', fontdict=font2, x=0.8, y=1.25, ha='center')
plt.tight_layout()

plt.show()
```

Bar Chart

```python
font = {'family': 'Courier New',
        'color':  'white',
        'weight': 'bold',
        'size': 40,
        }

categories = df_can['major_category'].unique()

totals = df_can.groupby('major_category').sum()
totals = totals['value'].to_numpy()

colors = ['lightblue','aquamarine','steelblue','royalblue','skyblue',
          'blue','cyan','turquoise','aqua']

fig, ax = plt.subplots(figsize=(20,10))
ax.set_facecolor("white")

ax.bar(categories, totals, color=colors, width=0.6)

ax.set_xlabel('Type of Crimes', fontsize=30, color='chocolate')
ax.set_ylabel('Number of Crimes', fontsize=30, color='sienna')

ax.tick_params(axis='both', which='major', labelsize=19, labelcolor='brown')
ax.set_xticklabels(categories, rotation=-45)
ax.yaxis.set_major_formatter('{x:,.0f}')

ax.set_title('Number of Crimes in Major Type\n', fontsize=30, color='black')

plt.show()
```

Scatter Plot

```python
font = {'family': 'Arial',
        'color':  'black',
        'weight': 'bold',
        'size': 40,
        }
year2016=df_can.loc[df_can['year'] == 2016]

months=year2016['month'].unique()

months.sort()

sum_month=year2016.groupby(['month']).sum()

Values=sum_month['value'].to_numpy()
name_of_months=['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']
print("no of crime in each month in the year 2016")
print(name_of_months)
print(Values)
n1 = np.arange(len(months))
xrange=range(1,13)
    
plt.figure(figsize=(20,10))
ax = plt.axes()
ax.set_facecolor("cyan")
plt.scatter(xrange,Values,s=250,color='brown',cmap='viridis',alpha=1)
plt.title("Crime in each month in the year 2016\n",fontdict=font)
plt.xlabel('Months',fontsize=30,color='deepskyblue')
plt.ylabel('Number of crimes',fontsize=30,color='darkcyan')
plt.xticks(n1,name_of_months)
plt.xticks(color='goldenrod',fontsize=20)
plt.yticks(color='goldenrod',fontsize=20)
```

Berdasarkan scatter plot, pada tahun 2016 bulan Juli merupakan bulan yang paling banyak melakukan kejahatan urutannya dimundurkan karena hasil output melenceng satu, dan yang paling sedikit melakukan kejahatan pada bulan febuari

Word Clouds

```python
stopwords = set(STOPWORDS)
```

cell tersebut digunakan untuk menggunakan wordcloud dengan stopwords yang berguna untuk mengurangi jumlah kata dalam sebuah dokumen

```python
borough = df_can.groupby('borough').size().reset_index(name='counts')

word_dict = dict(zip(borough['borough'].tolist(), borough['counts'].tolist()))

wordcloud = WordCloud(width = 800, height = 800, background_color ='white', min_font_size = 10).generate_from_frequencies(word_dict)

plt.figure(figsize = (8, 8), facecolor = None)
plt.imshow(wordcloud)
plt.axis("off")
plt.tight_layout(pad = 0)
  
plt.show()
```

cell yang digunakana adalah syntax untuk menggunakan wordcloud, berdasarkan borough, jika semakin besar kejahatan pada kota tersebut maka semakin besar pula nama kota yang tercetak
